````python
import ast

# 'annotation' 컬럼에서 감정(emotion) 값을 추출하는 함수 정의
def extract_emotions(annotation_string):
    try:
        # 문자열을 파이썬 리터럴 (리스트와 딕셔너리)로 안전하게 평가
        annotations = ast.literal_eval(annotation_string)
        # 각 항목에서 'emotion' 키의 값을 추출 (만약 'emotion' 키가 존재한다면)
        emotions = [item['emotion'] for item in annotations if 'emotion' in item]
        return emotions
    except (ValueError, SyntaxError):
        # 문자열이 유효한 리터럴이 아닌 경우 빈 리스트 반환
        return []

# 'annotation' 컬럼의 각 행에 함수를 적용하고 결과 리스트들을 하나의 리스트로 합침
all_emotions = df['annotation'].apply(extract_emotions).sum()

# 모든 감정 값들 중에서 고유한 값들만 추출하여 집합(set)으로 저장
unique_emotions = set(all_emotions)

# 고유한 감정 값 출력
display(unique_emotions)
````

````python
import ast
from collections import Counter

# 'annotation' 컬럼에서 감정(emotion) 값을 추출하는 함수 정의 (이전 함수 재사용)
def extract_emotions(annotation_string):
    try:
        # 문자열을 파이썬 리터럴 (리스트와 딕셔너리)로 안전하게 평가
        annotations = ast.literal_eval(annotation_string)
        # 각 항목에서 'emotion' 키의 값을 추출 (만약 'emotion' 키가 존재한다면)
        emotions = [item['emotion'] for item in annotations if 'emotion' in item]
        return emotions
    except (ValueError, SyntaxError):
        # 문자열이 유효한 리터럴이 아닌 경우 빈 리스트 반환
        return []

# 각 행을 순회하며 감정 개수 집계
for index, row in df.iterrows():
    context = row['context']
    emotions = extract_emotions(row['annotation'])

    # 현재 행의 감정 리스트에서 각 감정의 빈도를 계산
    emotion_counts = Counter(emotions)

    print(f"Context: {context}")
    print("Emotion Counts:")
    if emotion_counts:
        for emotion, count in emotion_counts.items():
            print(f"- {emotion}: {count}")
    else:
        print("해당 행에 감정이 없습니다.") # No emotions found for this row.
    print("-" * 20) # 구분선 출력
````

````python
import ast
from collections import Counter

# 감정을 추출하고 개수를 세는 함수 (이전 함수 재사용)
def count_emotions(annotation_string):
    try:
        annotations = ast.literal_eval(annotation_string)
        emotions = [item['emotion'] for item in annotations if 'emotion' in item]
        # 감정 개수를 딕셔너리로 반환
        return dict(Counter(emotions))
    except (ValueError, SyntaxError):
        return {} # 파싱이 실패하면 빈 딕셔너리 반환

# 'annotation' 컬럼에 함수를 적용하고 새로운 컬럼 'emotion_counts' 생성
df['emotion_counts'] = df['annotation'].apply(count_emotions)

# 새로운 컬럼이 포함된 처음 몇 행 표시
display(df.head())
````

# 전처리

컨텍스트와 감정이 1:1이 되도록 전처리

````python
# emotion_counts_df에서 최대값 감정 분석 및 전처리

import pandas as pd
from collections import Counter

def find_max_emotions(emotion_counts_dict):
    """
    emotion_counts 딕셔너리에서 최대값을 가진 감정들을 찾는 함수
    
    Args:
        emotion_counts_dict: {'감정1': 개수1, '감정2': 개수2, ...} 형태의 딕셔너리
    
    Returns:
        list: 최대값을 가진 감정들의 리스트
    """
    if not emotion_counts_dict:
        return []
    
    max_count = max(emotion_counts_dict.values())
    max_emotions = [emotion for emotion, count in emotion_counts_dict.items() if count == max_count]
    
    return max_emotions

def split_context_by_max_emotions(row):
    """
    각 행의 context를 최대값 감정들로 분할하는 함수
    
    Args:
        row: emotion_counts_df의 각 행
    
    Returns:
        list: 분할된 데이터의 리스트 (각 항목은 dict)
    """
    context = row['context']
    emotion_counts = row['emotion_counts']
    
    # 최대값 감정들 찾기
    max_emotions = find_max_emotions(emotion_counts)
    
    if not max_emotions:
        # 감정이 없는 경우 원본 그대로 반환
        return [{'context': context, 'emotion': None, 'emotion_count': 0}]
    
    # 각 최대값 감정에 대해 분할된 데이터 생성
    split_data = []
    for emotion in max_emotions:
        split_data.append({
            'context': context,
            'emotion': emotion,
            'emotion_count': emotion_counts[emotion],
            'original_emotion_counts': emotion_counts
        })
    
    return split_data

# 테스트를 위해 처음 5개 행에 대해 함수 적용
print("=== 최대값 감정 분석 테스트 ===")
for i in range(5):
    row = emotion_counts_df.iloc[i]
    max_emotions = find_max_emotions(row['emotion_counts'])
    print(f"Row {i}:")
    print(f"  Context: {row['context'][:50]}...")
    print(f"  Emotion counts: {row['emotion_counts']}")
    print(f"  Max emotions: {max_emotions}")
    print("-" * 50)
````

````python
# 전체 데이터에 대해 전처리 실행

def preprocess_emotion_data(emotion_counts_df):
    """
    emotion_counts_df를 전처리하여 최대값 감정으로 분류된 새로운 데이터프레임 생성
    
    Args:
        emotion_counts_df: 원본 데이터프레임
    
    Returns:
        pd.DataFrame: 전처리된 데이터프레임
    """
    processed_data = []
    
    for idx, row in emotion_counts_df.iterrows():
        # 각 행을 최대값 감정들로 분할
        split_data = split_context_by_max_emotions(row)
        
        # 분할된 데이터를 리스트에 추가
        for data in split_data:
            processed_data.append(data)
    
    # 새로운 데이터프레임 생성
    processed_df = pd.DataFrame(processed_data)
    
    return processed_df

# 전처리 실행
print("=== 전체 데이터 전처리 시작 ===")
processed_df = preprocess_emotion_data(emotion_counts_df)

print(f"원본 데이터 행 수: {len(emotion_counts_df)}")
print(f"전처리 후 데이터 행 수: {len(processed_df)}")
print(f"증가율: {len(processed_df) / len(emotion_counts_df):.2f}배")

print("\n=== 전처리 결과 샘플 ===")
print(processed_df.head(10))
````

````python
# 동점인 경우의 처리 결과 확인

print("=== 동점 감정이 있는 경우들 확인 ===")

# 동점이 있는 행들을 찾기
tie_cases = []
for idx, row in emotion_counts_df.iterrows():
    max_emotions = find_max_emotions(row['emotion_counts'])
    if len(max_emotions) > 1:  # 동점이 있는 경우
        tie_cases.append((idx, row, max_emotions))

print(f"동점이 있는 총 행 수: {len(tie_cases)}")

# 처음 5개 동점 케이스 상세 확인
for i, (idx, row, max_emotions) in enumerate(tie_cases[:5]):
    print(f"\n--- 동점 케이스 {i+1} (원본 행 {idx}) ---")
    print(f"Context: {row['context']}")
    print(f"Emotion counts: {row['emotion_counts']}")
    print(f"동점 감정들: {max_emotions}")
    
    # 전처리된 결과에서 해당 context의 분할 결과 확인
    split_results = processed_df[processed_df['context'] == row['context']]
    print(f"분할된 행 수: {len(split_results)}")
    for _, split_row in split_results.iterrows():
        print(f"  -> 감정: {split_row['emotion']}, 개수: {split_row['emotion_count']}")
    print("-" * 60)
````

````python
# 전처리 결과 통계 및 검증

print("=== 전처리 결과 통계 ===")

# 감정별 분포 확인
emotion_distribution = processed_df['emotion'].value_counts()
print(f"전처리 후 감정 종류 수: {len(emotion_distribution)}")
print(f"\n상위 10개 감정 분포:")
print(emotion_distribution.head(10))

# 동점으로 인한 데이터 증가 분석
single_emotion_cases = len(emotion_counts_df) - len(tie_cases)
multiple_emotion_cases = len(tie_cases)

print(f"\n=== 데이터 분할 분석 ===")
print(f"단일 최대값 감정 케이스: {single_emotion_cases}개")
print(f"동점 감정 케이스: {multiple_emotion_cases}개")
print(f"동점 케이스 비율: {multiple_emotion_cases / len(emotion_counts_df) * 100:.1f}%")

# 동점 케이스들의 평균 분할 수 계산
if tie_cases:
    total_splits = sum(len(find_max_emotions(row['emotion_counts'])) for _, row, _ in tie_cases)
    avg_splits = total_splits / len(tie_cases)
    print(f"동점 케이스 평균 분할 수: {avg_splits:.2f}개")

# 전처리된 데이터 저장
processed_df.to_excel("processed_emotion_data.xlsx", index=False)
print(f"\n전처리된 데이터가 'processed_emotion_data.xlsx'로 저장되었습니다.")
````

```
2. **표준 대조**: `label_standard.txt` 기준으로 표준 감정 확인
3. **데이터 개수 확인**: 각 감정별 데이터 개수 계산
4. **허용 기준 적용**: 
   - 표준 감정 (68개) + 25개 이상 데이터 감정
5. **데이터 필터링**: 허용 기준에 맞는 데이터만 유지
6. **결과 저장**: 정리된 데이터와 제거된 데이터를 별도 파일로 저장

## 결과 파일 ( datas 폴더 내부 )

- **`04_cleaned_emotion_data.xlsx`**: 정리된 최종 데이터
- **`04_removed_emotion_data.xlsx`**: 제거된 데이터 (참고용)

## 장점

- **표준 준수**: 라벨링 표준에 맞는 일관된 감정 분류
- **데이터 보존**: 충분한 데이터가 있는 비표준 감정도 유지
- **품질 향상**: 부족한 데이터로 인한 학습 저하 방지
- **투명성**: 제거된 데이터를 별도로 보관하여 추적 가능
```

````python
# label_standard.txt 기준으로 잘못된 감정 라벨 찾기

# 표준 감정 라벨 정의 (label_standard.txt 기준 + 중립 허용)
standard_emotions = {
    '기쁨': ['반가움', '즐거움', '신명남', '자신감', '감동', '만족감', '편안함', '고마움', '신뢰감', '안정감', '공감', '자랑스러움'],
    '사랑': ['호감', '귀중함', '매력적', '두근거림', '아른거림', '너그러움', '열정적임', '다정함', '동정(슬픔)'],
    '슬픔': ['억울함', '외로움', '후회', '실망', '허망', '그리움', '수치심', '고통', '절망', '무기력', '아픔'],
    '두려움': ['위축감', '놀람', '공포', '걱정', '초조함'],
    '분노': ['원망', '불쾌', '날카로움', '타오름'],
    '미움(상대방)': ['반감', '경멸', '비위상함', '치사함', '불신감', '시기심', '외면', '냉담'],
    '욕망': ['욕심', '궁금함', '아쉬움', '불만', '갈등', '기대감'],
    '싫어함(상태)': ['답답함', '불편함', '난처함', '서먹함', '심심함', '싫증'],
    '수치심': ['부끄러움', '죄책감', '미안함'],
    '중립': ['중립', '무감정', '평범함', '보통', '일반적']  # 중립 중분류 추가
}

# 모든 표준 감정 라벨을 하나의 리스트로 만들기
all_standard_emotions = []
for main_category, sub_emotions in standard_emotions.items():
    for sub_emotion in sub_emotions:
        all_standard_emotions.append(f"{main_category}_{sub_emotion}")

print(f"표준 감정 라벨 총 개수: {len(all_standard_emotions)}")
print(f"표준 감정 라벨들: {all_standard_emotions[:10]}...")  # 처음 10개만 출력
print(f"중립 감정 라벨들: {[emotion for emotion in all_standard_emotions if emotion.startswith('중립_')]}")

# 전처리된 데이터에서 실제 사용된 감정 라벨들 확인
actual_emotions = processed_df['emotion'].dropna().unique()
print(f"\n실제 데이터에서 사용된 감정 라벨 개수: {len(actual_emotions)}")

# 표준에 없는 잘못된 감정 라벨들 찾기
invalid_emotions = []
for emotion in actual_emotions:
    if emotion not in all_standard_emotions:
        invalid_emotions.append(emotion)

print(f"\n표준에 없는 잘못된 감정 라벨 개수: {len(invalid_emotions)}")
print("잘못된 감정 라벨들:")
for emotion in sorted(invalid_emotions):
    print(f"  - {emotion}")
````

````python
# 잘못된 감정 라벨들의 상세 분석

print("=== 잘못된 감정 라벨 상세 분석 ===")

# 잘못된 감정 라벨별 데이터 개수 확인
invalid_emotion_counts = processed_df[processed_df['emotion'].isin(invalid_emotions)]['emotion'].value_counts()

print("잘못된 감정 라벨별 데이터 개수:")
for emotion, count in invalid_emotion_counts.items():
    print(f"  {emotion}: {count}개")

# 잘못된 감정 라벨이 포함된 총 데이터 개수
total_invalid_data = len(processed_df[processed_df['emotion'].isin(invalid_emotions)])
total_data = len(processed_df)

print(f"\n잘못된 감정 라벨이 포함된 총 데이터 개수: {total_invalid_data}개")
print(f"전체 데이터 대비 비율: {total_invalid_data / total_data * 100:.2f}%")

# 잘못된 감정 라벨들의 패턴 분석
print(f"\n=== 잘못된 감정 라벨 패턴 분석 ===")

# 중분류별로 잘못된 라벨들 그룹화
invalid_by_category = {}
for emotion in invalid_emotions:
    main_category = emotion.split('_')[0] if '_' in emotion else emotion
    if main_category not in invalid_by_category:
        invalid_by_category[main_category] = []
    invalid_by_category[main_category].append(emotion)

print("중분류별 잘못된 감정 라벨들:")
for category, emotions in invalid_by_category.items():
    print(f"  {category}: {emotions}")

# 표준에 있는 중분류와 비교
print(f"\n표준 중분류: {list(standard_emotions.keys())}")
print(f"잘못된 라벨의 중분류: {list(invalid_by_category.keys())}")

# 표준에 없는 중분류 찾기 (중립 제외)
non_standard_categories = set(invalid_by_category.keys()) - set(standard_emotions.keys())
if non_standard_categories:
    print(f"\n표준에 없는 중분류: {list(non_standard_categories)}")
else:
    print(f"\n모든 잘못된 라벨의 중분류는 표준에 존재함")

# 중립 관련 라벨 확인
neutral_emotions = [emotion for emotion in invalid_emotions if emotion.startswith('중립_')]
if neutral_emotions:
    print(f"\n중립 관련 잘못된 라벨들: {neutral_emotions}")
    print("(이제 중립이 허용되므로 이들은 정리된 데이터에 포함될 예정)")
````

````python
# 잘못된 감정 라벨 제거 및 정리된 데이터 생성

print("=== 잘못된 감정 라벨 제거 (표준 + 25개 이상 데이터) ===")

# 각 감정별 데이터 개수 확인
emotion_counts = processed_df['emotion'].value_counts()
print(f"감정별 데이터 개수 (상위 20개):")
print(emotion_counts.head(20))

# 25개 이상인 감정 라벨들 찾기
sufficient_data_emotions = emotion_counts[emotion_counts >= 25].index.tolist()
print(f"\n25개 이상 데이터를 가진 감정 라벨 수: {len(sufficient_data_emotions)}")
print(f"25개 이상 감정 라벨들: {sufficient_data_emotions}")

# 허용할 감정 라벨들 (표준 + 25개 이상)
allowed_emotions = set(all_standard_emotions) | set(sufficient_data_emotions)
print(f"\n허용할 총 감정 라벨 수: {len(allowed_emotions)}")

# 허용할 감정 라벨만 포함하는 데이터프레임 생성
cleaned_df = processed_df[processed_df['emotion'].isin(allowed_emotions)].copy()

print(f"\n원본 전처리 데이터: {len(processed_df)}개")
print(f"정리된 데이터: {len(cleaned_df)}개")
print(f"제거된 데이터: {len(processed_df) - len(cleaned_df)}개")
print(f"제거 비율: {(len(processed_df) - len(cleaned_df)) / len(processed_df) * 100:.2f}%")

# 제거된 감정 라벨들 분석
removed_emotions = set(processed_df['emotion'].unique()) - allowed_emotions
print(f"\n제거된 감정 라벨 수: {len(removed_emotions)}")
if removed_emotions:
    print("제거된 감정 라벨들:")
    for emotion in sorted(removed_emotions):
        count = emotion_counts.get(emotion, 0)
        print(f"  - {emotion}: {count}개")

# 정리된 데이터의 감정 분포 확인
print(f"\n=== 정리된 데이터의 감정 분포 ===")
cleaned_emotion_distribution = cleaned_df['emotion'].value_counts()
print(f"정리된 데이터의 감정 종류 수: {len(cleaned_emotion_distribution)}")
print(f"\n상위 10개 감정 분포:")
print(cleaned_emotion_distribution.head(10))

# 중분류별 분포 확인
print(f"\n=== 중분류별 분포 ===")
cleaned_df['main_category'] = cleaned_df['emotion'].apply(lambda x: x.split('_')[0] if '_' in x else x)
main_category_distribution = cleaned_df['main_category'].value_counts()
print(main_category_distribution)

# 정리된 데이터 저장
cleaned_df.to_excel("cleaned_emotion_data.xlsx", index=False)
print(f"\n정리된 데이터가 'cleaned_emotion_data.xlsx'로 저장되었습니다.")

# 제거된 데이터도 별도로 저장 (참고용)
removed_data = processed_df[processed_df['emotion'].isin(removed_emotions)]
removed_data.to_excel("removed_emotion_data.xlsx", index=False)
print(f"제거된 데이터가 'removed_emotion_data.xlsx'로 저장되었습니다.")

# 25개 이상으로 살린 비표준 감정들 확인
non_standard_kept = set(sufficient_data_emotions) - set(all_standard_emotions)
if non_standard_kept:
    print(f"\n25개 이상 데이터로 살린 비표준 감정들: {len(non_standard_kept)}개")
    for emotion in sorted(non_standard_kept):
        count = emotion_counts.get(emotion, 0)
        print(f"  - {emotion}: {count}개")
````

````python
# 잘못된 감정 라벨의 구체적인 예시 확인

print("=== 잘못된 감정 라벨 구체적 예시 ===")

# 잘못된 감정 라벨이 포함된 데이터의 샘플 확인
print("잘못된 감정 라벨이 포함된 데이터 샘플:")
invalid_samples = processed_df[processed_df['emotion'].isin(invalid_emotions)].head(10)

for idx, row in invalid_samples.iterrows():
    print(f"\n--- 샘플 {idx} ---")
    print(f"Context: {row['context']}")
    print(f"잘못된 감정: {row['emotion']}")
    print(f"감정 개수: {row['emotion_count']}")
    print(f"원본 감정 카운트: {row['original_emotion_counts']}")

# 표준과 비슷한 잘못된 라벨들 찾기 (오타나 유사한 표현)
print(f"\n=== 표준과 유사한 잘못된 라벨 분석 ===")

def find_similar_emotions(invalid_emotion, standard_emotions):
    """잘못된 감정과 유사한 표준 감정 찾기"""
    similar = []
    for standard in standard_emotions:
        if invalid_emotion in standard or standard in invalid_emotion:
            similar.append(standard)
    return similar

print("잘못된 감정과 유사한 표준 감정들:")
for invalid_emotion in sorted(invalid_emotions)[:10]:  # 처음 10개만 확인
    similar = find_similar_emotions(invalid_emotion, all_standard_emotions)
    if similar:
        print(f"  {invalid_emotion} -> 유사한 표준: {similar}")
    else:
        print(f"  {invalid_emotion} -> 유사한 표준 없음")

print(f"\n=== 최종 요약 ===")
print(f"✅ 표준 감정 라벨 총 개수: {len(all_standard_emotions)}개 (중립 포함)")
print(f"✅ 25개 이상 데이터 감정 라벨: {len(sufficient_data_emotions)}개")
print(f"✅ 허용된 총 감정 라벨: {len(allowed_emotions)}개")
print(f"✅ 실제 데이터의 감정 라벨: {len(actual_emotions)}개")
print(f"✅ 정리된 데이터: {len(cleaned_df)}개")
print(f"✅ 제거된 데이터: {len(processed_df) - len(cleaned_df)}개")
print(f"✅ 정리된 데이터 파일: 'cleaned_emotion_data.xlsx'")
print(f"✅ 제거된 데이터 파일: 'removed_emotion_data.xlsx'")
print(f"✅ 중립 감정 허용: 중립_중립, 중립_무감정, 중립_평범함, 중립_보통, 중립_일반적")
print(f"✅ 25개 이상 데이터로 살린 비표준 감정: {len(non_standard_kept)}개")
````
