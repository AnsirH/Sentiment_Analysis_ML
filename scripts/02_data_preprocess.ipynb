```python
# 데이터 로드
import pandas as pd

# 상대 경로 사용
df = pd.read_excel('../datas/01_dataframe.xlsx')

print(f"데이터 로드 완료: {len(df)}개 행")
print(f"컬럼: {list(df.columns)}")
print(f"\n데이터 미리보기:")
print(df.head())
```

```python
# 불필요 컬럼 제거
emotion_counts_df = df.copy()
emotion_counts_df.drop(['created_date', 'source', '__index_level_0__'], axis=1, inplace=True)
```

```python
print(f"\n데이터 미리보기:")
display(emotion_counts_df.head())
```

```python
# annotation 컬럼에서 감정 추출 및 개수 세기
# annotation 컬럼 제거
import ast
from collections import Counter

# 감정을 추출하고 개수를 세는 함수
def count_emotions(annotation_string):
    try:
        annotations = ast.literal_eval(annotation_string)
        emotions = [item['emotion'] for item in annotations if 'emotion' in item]
        # 감정 개수를 딕셔너리로 반환
        return dict(Counter(emotions))
    except (ValueError, SyntaxError):
        return {} # 파싱이 실패하면 빈 딕셔너리 반환

# 'annotation' 컬럼에 함수를 적용하고 새로운 컬럼 'emotion_counts' 생성
emotion_counts_df['emotion_counts'] = emotion_counts_df['annotation'].apply(count_emotions)

# annotation 컬럼 제거
emotion_counts_df.drop('annotation', axis=1, inplace=True)

# 새로운 컬럼이 포함된 처음 몇 행 표시
display(emotion_counts_df.head())
```

```python
standard_emotions = {
    '기쁨': ['반가움', '즐거움', '신명남', '자신감', '감동', '만족감', '편안함', '고마움', '신뢰감', '안정감', '공감', '자랑스러움'],
    '사랑': ['호감', '귀중함', '매력적', '두근거림', '아른거림', '너그러움', '열정적임', '다정함', '동정(슬픔)'],
    '슬픔': ['억울함', '외로움', '후회', '실망', '허망', '그리움', '수치심', '고통', '절망', '무기력', '아픔'],
    '두려움': ['위축감', '놀람', '공포', '걱정', '초조함'],
    '분노': ['원망', '불쾌', '날카로움', '타오름'],
    '미움(상대방)': ['반감', '경멸', '비위상함', '치사함', '불신감', '시기심', '외면', '냉담'],
    '욕망': ['욕심', '궁금함', '아쉬움', '불만', '갈등', '기대감'],
    '싫어함(상태)': ['답답함', '불편함', '난처함', '서먹함', '심심함', '싫증'],
    '수치심': ['부끄러움', '죄책감', '미안함']
}
```

```python
# emotion_counts_df에서 실제 사용된 감정 라벨들 추출
print("=== 실제 사용된 감정 라벨 추출 ===")

# emotion_counts 컬럼에서 모든 감정 라벨을 추출
actual_emotions = set()
for emotion_counts_dict in emotion_counts_df['emotion_counts']:
    if isinstance(emotion_counts_dict, dict):
        actual_emotions.update(emotion_counts_dict.keys())

print(f"실제 데이터에서 사용된 감정 라벨 개수: {len(actual_emotions)}")
print(f"실제 사용된 감정 라벨들( 초반 10개 ):")
for emotion in sorted(actual_emotions)[:10]:
    print(f"  {emotion}")

# 각 감정별 총 개수 계산
emotion_total_counts = Counter()
for emotion_counts_dict in emotion_counts_df['emotion_counts']:
    if isinstance(emotion_counts_dict, dict):
        emotion_total_counts.update(emotion_counts_dict)

print(f"\n감정별 총 개수 (상위 10개):")
for emotion, count in emotion_total_counts.most_common(10):
    print(f"  {emotion}: {count}개")
```

```python
# actual_emotions에서 중립 관련 감정들 확인
print("=== actual_emotions에서 중립 관련 감정 확인 ===")

neutral_emotions_in_data = [emotion for emotion in actual_emotions if emotion.startswith('중립_')]
print(f"실제 데이터에 있는 중립 감정들: {neutral_emotions_in_data}")

# 중립 중분류의 소분류들 추출
neutral_subcategories = set()
for emotion in neutral_emotions_in_data:
    if '_' in emotion:
        subcategory = emotion.split('_', 1)[1]  # '중립_' 이후 부분
        neutral_subcategories.add(subcategory)

print(f"중립 중분류의 소분류들: {sorted(neutral_subcategories)}")

# standard_emotions에 중립 추가
standard_emotions['중립'] = sorted(neutral_subcategories)

print(f"\n업데이트된 standard_emotions:")
for category, subcategories in standard_emotions.items():
    print(f"  {category}: {subcategories}")

# 모든 표준 감정 라벨을 다시 생성
all_standard_emotions = []
for main_category, sub_emotions in standard_emotions.items():
    for sub_emotion in sub_emotions:
        all_standard_emotions.append(f"{main_category}_{sub_emotion}")

print(f"\n업데이트된 표준 감정 라벨 총 개수: {len(all_standard_emotions)}")
print(f"중립 감정 라벨들: {[emotion for emotion in all_standard_emotions if emotion.startswith('중립_')]}")
```

```python
# 기준에 없고, 중립도 아니고, 25개 미만인 데이터 추출
print("=== 제거할 감정 라벨 식별 ===")

# 허용할 감정 라벨들 정의
# 1. 표준 감정 라벨들
allowed_emotions = set(all_standard_emotions)

# 2. 25개 이상 데이터가 있는 감정 라벨들 추가
for emotion, count in emotion_total_counts.items():
    if count >= 25:
        allowed_emotions.add(emotion)

print(f"허용할 감정 라벨 총 개수: {len(allowed_emotions)}")

# 제거할 감정 라벨들 (기준에 없고, 중립도 아니고, 25개 미만)
to_remove_emotions = set(actual_emotions) - allowed_emotions

print(f"\n제거할 감정 라벨 개수: {len(to_remove_emotions)}")
print(f"제거할 감정 라벨들:")
for emotion in sorted(to_remove_emotions):
    count = emotion_total_counts[emotion]
    print(f"  {emotion}: {count}개")

# 제거할 데이터의 총 개수
total_removed_count = sum(emotion_total_counts[emotion] for emotion in to_remove_emotions)
total_data_count = sum(emotion_total_counts.values())

print(f"\n제거할 데이터 총 개수: {total_removed_count}개")
print(f"전체 데이터 대비 제거 비율: {total_removed_count / total_data_count * 100:.2f}%")
```

```python
# 허용할 감정 라벨과 제거할 감정 라벨 정보를 엑셀 파일로 저장
print("=== 감정 라벨 정보 엑셀 저장 ===")

# 허용할 감정 라벨 정보 생성
allowed_emotions_info = []
for emotion in sorted(allowed_emotions):
    count = emotion_total_counts.get(emotion, 0)
    category = emotion.split('_')[0] if '_' in emotion else emotion
    subcategory = emotion.split('_', 1)[1] if '_' in emotion else emotion
    
    # 표준 여부 확인
    is_standard = emotion in all_standard_emotions
    # 25개 이상 여부 확인
    is_sufficient_data = count >= 25
    
    allowed_emotions_info.append({
        'emotion': emotion,
        'category': category,
        'subcategory': subcategory,
        'count': count,
        'is_standard': is_standard,
        'is_sufficient_data': is_sufficient_data,
        'reason': '표준' if is_standard else ('25개 이상' if is_sufficient_data else '기타')
    })

# 제거할 감정 라벨 정보 생성
removed_emotions_info = []
for emotion in sorted(to_remove_emotions):
    count = emotion_total_counts.get(emotion, 0)
    category = emotion.split('_')[0] if '_' in emotion else emotion
    subcategory = emotion.split('_', 1)[1] if '_' in emotion else emotion
    
    # 제거 이유 분석
    if emotion in all_standard_emotions:
        reason = '표준에 있음 (오류)'
    elif count >= 25:
        reason = '25개 이상 (오류)'
    else:
        reason = '표준에 없고 25개 미만'
    
    removed_emotions_info.append({
        'emotion': emotion,
        'category': category,
        'subcategory': subcategory,
        'count': count,
        'reason': reason
    })

# DataFrame 생성
allowed_df = pd.DataFrame(allowed_emotions_info)
removed_df = pd.DataFrame(removed_emotions_info)

# 엑셀 파일로 저장 (여러 시트)
with pd.ExcelWriter('../datas/emotion_label_analysis.xlsx', engine='openpyxl') as writer:
    # 허용할 감정 라벨
    allowed_df.to_excel(writer, sheet_name='허용할_감정_라벨', index=False)
    
    # 제거할 감정 라벨
    removed_df.to_excel(writer, sheet_name='제거할_감정_라벨', index=False)
    
    # 요약 통계
    summary_data = {
        '항목': [
            '전체 감정 라벨 종류 수',
            '허용할 감정 라벨 종류 수',
            '제거할 감정 라벨 종류 수',
            '전체 감정 데이터 총 개수',
            '허용할 감정 데이터 총 개수',
            '제거할 감정 데이터 총 개수',
            '제거 비율 (%)'
        ],
        '값': [
            len(actual_emotions),
            len(allowed_emotions),
            len(to_remove_emotions),
            sum(emotion_total_counts.values()),
            sum(emotion_total_counts[emotion] for emotion in allowed_emotions),
            sum(emotion_total_counts[emotion] for emotion in to_remove_emotions),
            round(sum(emotion_total_counts[emotion] for emotion in to_remove_emotions) / sum(emotion_total_counts.values()) * 100, 2)
        ]
    }
    summary_df = pd.DataFrame(summary_data)
    summary_df.to_excel(writer, sheet_name='요약_통계', index=False)

print(f"✅ 허용할 감정 라벨 종류 수: {len(allowed_emotions)}")
print(f"✅ 제거할 감정 라벨 종류 수: {len(to_remove_emotions)}")
print(f"✅ 허용할 감정 데이터 총 개수: {sum(emotion_total_counts[emotion] for emotion in allowed_emotions)}")
print(f"✅ 제거할 감정 데이터 총 개수: {sum(emotion_total_counts[emotion] for emotion in to_remove_emotions)}")
print(f"✅ 엑셀 파일 저장 완료: 'emotion_label_analysis.xlsx'")
print(f"   - 시트1: 허용할_감정_라벨")
print(f"   - 시트2: 제거할_감정_라벨") 
print(f"   - 시트3: 요약_통계")
```

```python
# 제거할 감정 라벨만 emotion_counts에서 제거
print("=== 제거할 감정 라벨만 제거 ===")

# 원본 데이터를 복사하여 수정
cleaned_emotion_counts_df = emotion_counts_df.copy()

# 제거할 감정 라벨이 포함된 행들 추적
rows_with_removed_emotions = []
total_removed_emotions = 0

for idx, row in cleaned_emotion_counts_df.iterrows():
    emotion_counts_dict = row['emotion_counts']
    if isinstance(emotion_counts_dict, dict):
        # 이 행에서 제거될 감정들 찾기
        removable_in_this_row = [emotion for emotion in emotion_counts_dict.keys() if emotion in to_remove_emotions]
        
        if removable_in_this_row:
            # 제거할 감정들을 emotion_counts에서 제거
            cleaned_emotion_counts = {k: v for k, v in emotion_counts_dict.items() if k not in to_remove_emotions}
            cleaned_emotion_counts_df.at[idx, 'emotion_counts'] = cleaned_emotion_counts
            
            rows_with_removed_emotions.append({
                'row_idx': idx,
                'context': row['context'],
                'original_emotion_counts': emotion_counts_dict,
                'cleaned_emotion_counts': cleaned_emotion_counts,
                'removed_emotions': removable_in_this_row
            })
            
            total_removed_emotions += len(removable_in_this_row)

print(f"감정 라벨이 제거된 행 개수: {len(rows_with_removed_emotions)}")
print(f"제거된 총 감정 라벨 개수: {total_removed_emotions}")

# 제거된 데이터 샘플 확인
print(f"\n제거된 데이터 샘플 (처음 5개):")
for i, sample in enumerate(rows_with_removed_emotions[:5]):
    print(f"\n--- 샘플 {i+1} (행 {sample['row_idx']}) ---")
    print(f"Context: {sample['context'][:100]}...")
    print(f"원본 Emotion counts: {sample['original_emotion_counts']}")
    print(f"정리된 Emotion counts: {sample['cleaned_emotion_counts']}")
    print(f"제거된 감정들: {sample['removed_emotions']}")

# 빈 emotion_counts가 된 행들 찾기 및 삭제
print(f"\n=== 빈 emotion_counts 행 삭제 ===")
empty_rows = []
for idx, row in cleaned_emotion_counts_df.iterrows():
    if isinstance(row['emotion_counts'], dict) and len(row['emotion_counts']) == 0:
        empty_rows.append(idx)

print(f"빈 emotion_counts가 된 행 개수: {len(empty_rows)}")

if empty_rows:
    print(f"삭제될 빈 행들 (처음 5개):")
    for i, idx in enumerate(empty_rows[:5]):
        print(f"  행 {idx}: {cleaned_emotion_counts_df.iloc[idx]['context'][:50]}...")
    
    # 빈 행들 삭제
    cleaned_emotion_counts_df = cleaned_emotion_counts_df.drop(empty_rows).reset_index(drop=True)
    print(f"빈 행 {len(empty_rows)}개가 삭제되었습니다.")
else:
    print("삭제할 빈 행이 없습니다.")

print(f"\n최종 데이터 행 수: {len(cleaned_emotion_counts_df)}")

# 최종 정리된 데이터 저장
cleaned_emotion_counts_df.to_excel('../datas/06_processed_emotion_data.xlsx', index=False)
print(f"최종 정리된 데이터가 '06_processed_emotion_data.xlsx'로 저장되었습니다.")
```

```python
# 정리된 데이터의 최종 통계
print("=== 정리된 데이터 최종 통계 ===")

# 정리된 데이터에서 감정 라벨 분포 확인
final_emotion_counts = Counter()
for emotion_counts_dict in cleaned_emotion_counts_df['emotion_counts']:
    if isinstance(emotion_counts_dict, dict):
        final_emotion_counts.update(emotion_counts_dict)

print(f"정리 후 감정 종류 수: {len(final_emotion_counts)}")
print(f"정리 후 총 감정 데이터 개수: {sum(final_emotion_counts.values())}")

print(f"\n정리된 데이터의 상위 10개 감정 분포:")
for emotion, count in final_emotion_counts.most_common(10):
    print(f"  {emotion}: {count}개")

# 제거된 감정 라벨들의 통계
print(f"\n제거된 감정 라벨 통계:")
removed_emotion_stats = Counter()
for sample in rows_with_removed_emotions:
    for emotion in sample['removed_emotions']:
        removed_emotion_stats[emotion] += 1

print(f"제거된 감정 라벨 종류 수: {len(removed_emotion_stats)}")
for emotion, count in removed_emotion_stats.most_common(10):
    print(f"  {emotion}: {count}개 행에서 제거")


# 제거된 행들의 상세 정보 출력
print(f"\n=== 제거된 행 상세 정보 ===")
print(f"빈 emotion_counts로 인해 삭제된 행 개수: {len(empty_rows)}")

if empty_rows:
    print(f"삭제된 행들의 샘플 (처음 5개):")
    for i, idx in enumerate(empty_rows[:5]):
        # 원본 데이터에서 해당 행의 정보 가져오기
        original_row = emotion_counts_df.iloc[idx]
        print(f"\n--- 삭제된 행 {i+1} (원본 행 {idx}) ---")
        print(f"Context: {original_row['context'][:100]}...")
        print(f"원본 Emotion counts: {original_row['emotion_counts']}")
        
        # 이 행에서 제거된 감정들
        removed_in_this_row = [emotion for emotion in original_row['emotion_counts'].keys() if emotion in to_remove_emotions]
        print(f"제거된 감정들: {removed_in_this_row}")
        print(f"제거 후 남은 감정들: {[emotion for emotion in original_row['emotion_counts'].keys() if emotion not in to_remove_emotions]}")
else:
    print("삭제된 행이 없습니다.")

# 최종 요약 업데이트
print(f"\n=== 최종 요약 (업데이트) ===")
print(f"✅ 원본 데이터 행 수: {len(emotion_counts_df)}")
print(f"✅ 정리된 데이터 행 수: {len(cleaned_emotion_counts_df)}")
print(f"✅ 감정 라벨이 제거된 행 수: {len(rows_with_removed_emotions)}")
print(f"✅ 제거된 총 감정 라벨 개수: {total_removed_emotions}")
print(f"✅ 빈 emotion_counts로 삭제된 행 수: {len(empty_rows)}")
print(f"✅ 정리 후 감정 종류 수: {len(final_emotion_counts)}")
print(f"✅ 정리된 데이터 파일: '06_processed_emotion_data.xlsx'")
```
